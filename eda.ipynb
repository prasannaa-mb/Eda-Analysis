```python
merged = pd.merge(left=df,right=df_country,how="left",on="Country Code")
```

```python
merged.columns
```

```python
final_df = merged 
final_df.head(2)
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index 
country_values = final_df.Country.value_counts().values
plt.pie(country_values[:3],labels=country_names[:3])
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating_count"})
```

```python
#Observations
# 1. Rating color is based on Aggregate rating
# 2. Rating text is based on Aggregate rating   
# 3. Rating count is based on Aggregate rating, Rating color and Rating text

final_df.columns 
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating_Counts"})
ratings
```

```python
ratings.head()
```

```python
plt.figure(figsize=(12,3))
sns.barplot(x="Aggregate rating",y="Rating_Counts",data=ratings)
plt.show
```
